```python
## DATSET SPLITTING AND FEATURE SELECTION
```

```python
# Slice the DataFrame to select the desired range of data
df_training = df.iloc[index1:index2]
features = ['Ecell_V','I_mA', 'Temperature__C']
X = df_training.loc[:, features]
y = df_training.loc[:, ['SOC%']]
df_training.shape
```

```python
# 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=1)
```

```python
# MODELING

# 1. OLS

ols = LinearRegression()
ols.fit(X_train, y_train)
ols_yhat_test = ols.predict(X_test)
ols_yhat_val = ols.predict(X_val)
```

```python
# 1. TEST

print(cl('R-SQUARED:', attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of OLS model is {}'.format(r2(y_test, ols_yhat_test)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')

# 2. Validation

print(cl('R-SQUARED:', attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of OLS model is {}'.format(r2(y_val, ols_yhat_val)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')

```

```python
# MODELING

# create regressor object
regressor = RandomForestRegressor()
 
# fit the regressor with x and y data
regressor.fit(X_train, y_train) 
clf_yhat_test = regressor.predict(X_test)
clf_yhat_val = regressor.predict(X_val)

```

```python
# 1. TEST

print('-------------------------------------------------------------------------------')
print(cl('R-Squared of RandomForest model is {}'.format(r2(y_test, clf_yhat_test)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

```python
# 2. VAL

print('-------------------------------------------------------------------------------')
print(cl('R-Squared of Random Forest model is {}'.format(r2(y_val, clf_yhat_val)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

```python
# MODELLING


# Instantiate Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators = 32, max_depth = 1, random_state = 1)
  
# Fit to training set
gbr.fit(X_train, y_train)
  
# Predict on test set
gbr_yhat_test= gbr.predict(X_test)

# Predict on val set
gbr_yhat_val= gbr.predict(X_val)
```

```python
# 2. R-squared

print('-------------------------------------------------------------------------------')
print(cl('R-Squared of gbr model is {}'.format(r2(y_test, gbr_yhat_test)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

```python
# 2. R-squared

print('-------------------------------------------------------------------------------')
print(cl('R-Squared of gbr model is {}'.format(r2(y_val, gbr_yhat_val)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(128,32),max_iter=100, random_state=1)

# Train the model on the training data
mlp.fit(X_train, y_train)

```

```python
# Evaluate the model on the test data
test_score = mlp.score(X_test, y_test)
test_score
```

```python
# Evaluate the model on the validation data
val_score = mlp.score(X_val, y_val)
val_score
```
